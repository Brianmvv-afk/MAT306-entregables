Comenzamos leyendo los archivos descargados desde kaggle, los cuales no son soportados por github por sus tamaños.

```python
import pandas as pd
import os

def cargar_datos():
      distritos = pd.read_csv("learnplatform-covid19-impact-on-digital-learning/districts_info.csv")
      productos = pd.read_csv("learnplatform-covid19-impact-on-digital-learning/products_info.csv")
      engagement_path = "learnplatform-covid19-impact-on-digital-learning/engagement_data"
      archivos_engagement = []

      for archivo in os.listdir(engagement_path):
          if archivo.endswith('.csv'):
              district_id = int(archivo.split('.')[0])
              df_temp = pd.read_csv(f'{engagement_path}/{archivo}')
              df_temp['district_id'] = district_id
              archivos_engagement.append(df_temp)

      engagement = pd.concat(archivos_engagement, ignore_index=True)

      return distritos, productos, engagement

distritos_df, productos_df, engagement_df = cargar_datos()

print("Forma de los datasets:")
print(f"Distritos: {distritos_df.shape}")
print(f"Productos: {productos_df.shape}")
print(f"Engagement: {engagement_df.shape}")
```

Notamos en particular que, el dataset de *engagement* es muy grande (con 22 millones de entradas aprox). A continuación, creamos una funcióon para limpiar estos datos.

```python
def limpiar_datos(distritos, productos, engagement):
    distritos_clean = distritos.dropna(subset=['state']).copy()

    productos_clean = productos.dropna(subset=['Sector(s)']).copy()

    engagement_clean = engagement.dropna().copy()
    engagement_clean['time'] = pd.to_datetime(engagement_clean['time'])
    engagement_clean['month'] = engagement_clean['time'].dt.month
    engagement_clean['week'] = engagement_clean['time'].dt.isocalendar().week

    return distritos_clean, productos_clean, engagement_clean

distritos_clean, productos_clean, engagement_clean = limpiar_datos(distritos_df, productos_df, engagement_df)
print("Dimensiones después de limpiar:")
print(f"- distritos_clean: {distritos_clean.shape}")
print(f"- productos_clean: {productos_clean.shape}")
print(f"- engagement_clean: {engagement_clean.shape}\n")
```

Notamos como hubo una gran reducción de datos tras la limpieza para engagement_clean. Ahora vamos a estudiar la evolución promedio mensual del `engagement_index` y del `pct_access`. Para ello vamos agregando por mes (ya extraído en `engagement_clean['month']`) y graficamos la serie mensual de ambas métricas para identificar tendencias, picos, y caídas relevantes.

```python
import matplotlib.pyplot as plt

def analizar_engagement_temporal(engagement_df):

    engagement_mensual = engagement_df.groupby('month').agg({
        'engagement_index': 'mean',
        'pct_access': 'mean'
    }).reset_index()

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))
    ax1.plot(engagement_mensual['month'], engagement_mensual['engagement_index'],
             marker='o', linewidth=2, markersize=8)
    ax1.set_title('Evolución del Engagement Index Mensual', fontsize=16, fontweight='bold')
    ax1.set_xlabel('Mes')
    ax1.set_ylabel('Engagement Index Promedio')
    ax1.grid(True, alpha=0.3)

    ax2.plot(engagement_mensual['month'], engagement_mensual['pct_access'],
             marker='s', linewidth=2, markersize=8, color='orange')
    ax2.set_title('Evolución del Porcentaje de Acceso Mensual', fontsize=16, fontweight='bold')
    ax2.set_xlabel('Mes')
    ax2.set_ylabel('Porcentaje de Acceso Promedio')
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    return engagement_mensual

engagement_mensual = analizar_engagement_temporal(engagement_clean)
display(engagement_mensual)
```

Podemos notar las siguientes observaciones para cada gráfico:

- **Engagement Index**:
  - Observamos un aumento en los primeros meses (mes 1 a 4) seguido por una caída pronunciada alrededor de los meses intermedios (meses 6 y 7).
  - Hay un pico marcado alrededor del mes 9 y luego una estabilización en niveles altos (meses 9–12) en comparación al inicio.
  - La caída intermedia podría ser causada por periodos vacacionales o a un efecto estacional, dado el fin del semestre.

- **Porcentaje de Acceso**:
  - Inicia en valores altos (aprox. 0.8), desciende hasta un mínimo en el mes 7 (aprox. 0.14) y presenta un repunte muy fuerte en el mes 9 (aprox. 0.8), seguido por una ligera disminución hacia el final del año.
  - El patrón de `pct_access` es coherente con el `engagement_index` —ambas métricas muestran una caída a mitad de periodo y un repunte posterior— lo que sugiere un comportamiento agregado consistente.

  En esta sección combinaremos la información de `engagement` y `distritos` para analizar cómo varía el uso (engagement index y pct_access) a nivel de **estado**.

```python
def analizar_por_estado(distritos, engagement):
    merged_data = engagement.merge(distritos, on='district_id', how='left')
    engagement_estado = merged_data.groupby('state').agg({
        'engagement_index': 'mean',
        'pct_access': 'mean'
    }).nlargest(10, 'engagement_index')

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
    engagement_estado['engagement_index'].plot(kind='bar', ax=ax1, color='skyblue')
    ax1.set_title('Top 10 Estados por Engagement Index', fontsize=16, fontweight='bold')
    ax1.set_ylabel('Engagement Index Promedio')
    ax1.tick_params(axis='x', rotation=45)

    engagement_estado['pct_access'].plot(kind='bar', ax=ax2, color='lightcoral')
    ax2.set_title('Top 10 Estados por Porcentaje de Acceso', fontsize=16, fontweight='bold')
    ax2.set_ylabel('Porcentaje de Acceso Promedio')
    ax2.tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

    return merged_data

merged_data = analizar_por_estado(distritos_clean, engagement_clean)
```

A continuación, preparamos el modelo de Machine Learning. Considerando lo siguiente:
- Construiremos un dataset de observaciones agregadas a nivel **(district_id, month)** que sirva de unidad para el modelado.
- Incluiremos estadísticas resumen de engagement (`mean`, `std`, `sum`), métricas de acceso y características de `products` y `districts` (modo de categoría de producto, provider, etc.).

```python
def preparar_datos_ml(distritos, productos, engagement):
    datos_agregados = engagement.groupby(['district_id', 'month']).agg({
        'engagement_index': ['mean', 'std', 'sum'],
        'pct_access': ['mean', 'std'],
        'lp_id': 'nunique'
    }).reset_index()

    #"aplanamos" columnas multi-index
    datos_agregados.columns = ['_'.join(col).strip() if col[1] != '' else col[0]
                              for col in datos_agregados.columns.values]

    datos_ml = datos_agregados.merge(distritos, on='district_id', how='left')
    engagement_productos = engagement.merge(productos, left_on='lp_id', right_on='LP ID', how='left')
    productos_agregados = engagement_productos.groupby(['district_id', 'month']).agg({
        'Sector(s)': lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown',
        'Primary Essential Function': lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown',
        'Provider/Company Name': lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown'
    }).reset_index()

    datos_ml = datos_ml.merge(productos_agregados, on=['district_id', 'month'], how='left')
    return datos_ml

datos_ml = preparar_datos_ml(distritos_clean, productos_clean, engagement_clean)
print(f"Datos para ML: {datos_ml.shape}")
```

A continuación inspeccionamos la distribución del **target** que planteamos predecir, es decir, `engagement_index_mean`. Para determinar si una regresión u otro modelo es apropiado o debemos aplicar alguna transformación.

```python
import numpy as np

y = datos_ml['engagement_index_mean']
print("Resumen estadístico del target:")
print(y.describe())

plt.figure(figsize=(8,4))
plt.hist(y.dropna(), bins=50)
plt.title('Histograma de engagement_index_mean')
plt.show()
print("Skew:", y.dropna().skew())
```

Notamos que la variable `engagement_index_mean` está sesgada a la derecha (con skew 5.18), con una cola larga y algunos valores extremos (máximo cercano a 4.333). Además, la mediana (139.7) está bastante por debajo de la media (203), lo que confirma la asimetría.

Dado lo anterior, vamos a aplicar una transformación logarítmica al modelo RandomForest y HistGB para estabilizar la varianza y reducir el sesgo.

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import joblib


RND = 42 #fijamos semilla
datos = datos_ml.copy()

#definimos target y features
target_col = 'engagement_index_mean'
exclude_cols = ['district_id', 'engagement_index_mean', 'engagement_index_std',
                'engagement_index_sum', 'pct_access_std', 'pct_access_mean']

feature_cols = [c for c in datos.columns if c not in exclude_cols]
X = datos[feature_cols].copy()
y = datos[target_col].astype(float).copy()

print("Observaciones:", X.shape[0], " Features:", len(feature_cols))


#aplicamos un recorte a los outliers (para reducir influencia de la cola)
upper = y.quantile(0.99)
print(f"Clip en percentil 99 = {upper:.2f}")
y_clip = np.minimum(y, upper)

#transformación log1p
y_log = np.log1p(y_clip)
print("Skew antes:", y.skew(), "Skew después log1p:", pd.Series(y_log).skew())


num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()

# excluimos columnas que no aportan
for bad in ['Unnamed: 0']:
    if bad in num_cols:
        num_cols.remove(bad)
print("Num cols:", num_cols)
print("Cat cols:", cat_cols)

#aplicamos un preprocesador
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])
preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols)
], sparse_threshold=0)


X_train, X_test, y_train_log, y_test_log = train_test_split(
    X, y_log, test_size=0.25, random_state=RND
)
X_train_proc = preprocessor.fit_transform(X_train)
X_test_proc  = preprocessor.transform(X_test)
print("X_train_proc shape:", X_train_proc.shape)


rf = RandomForestRegressor(n_estimators=200, random_state=RND, n_jobs=-1)
rf.fit(X_train_proc, y_train_log)
hgb = HistGradientBoostingRegressor(random_state=RND)
hgb.fit(X_train_proc, y_train_log)


def eval_model(model, X_test_proc, y_test_log):
    preds_log = model.predict(X_test_proc)
    preds = np.expm1(preds_log)         # invertimos log1p
    y_true = np.expm1(y_test_log)
    mse = mean_squared_error(y_true, preds)
    rmse = np.sqrt(mse)
    mae  = mean_absolute_error(y_true, preds)
    r2   = r2_score(y_true, preds)
    return {'rmse': rmse, 'mae': mae, 'r2': r2, 'preds': preds, 'y_true': y_true}

rf_res = eval_model(rf, X_test_proc, y_test_log)
hgb_res = eval_model(hgb, X_test_proc, y_test_log)

print("RandomForest - RMSE:{rmse:.3f} MAE:{mae:.3f} R2:{r2:.3f}".format(**rf_res))
print("HistGB       - RMSE:{rmse:.3f} MAE:{mae:.3f} R2:{r2:.3f}".format(**hgb_res))

# conseguimos la importancia del os features
feature_names = list(num_cols) + list(cat_cols)
importances = rf.feature_importances_
imp_series = pd.Series(importances, index=feature_names).sort_values(ascending=False)
print("\nTop 5  features:")
display(imp_series.head(5))

plt.figure(figsize=(10,4))
imp_series.head(10).plot(kind='bar')
plt.title('Top 10 importancias')
plt.tight_layout()
plt.show()
```

Vemos como el modelo RandomForest supera a HistGB, explicando aproximadamente el 52% de la varianza del `engagement_index_mean` y tiene un error típico (RMSE) cercano a 132 puntos en la escala original. Dado que la media del target es 203 aprox., ese RMSE equivale a 65% de la media, por lo que, aunque el modelo capta señales relevantes, todavía presenta errores importantes, especialmente para observaciones en la cola alta.
El MAE (74.5) sugiere que el error absoluto típico es 75 puntos de engagement; la diferencia de RMSE vs MAE nos indica la presencia de outliers que elevan el RMSE.

Respectto a la importancia de las variables, `lp_id_nunique` (cantidad de productos distintos en un district-mes) es la variable más importante. Esto sugiere que la diversidad de herramientas utilizadas en un distrito está fuertemente asociada al engagement. Luego, `month` aparece como segunda variable, indicando un efecto temporal/estacional (vacaciones, periodos lectivos, etc.). Finalmente, `state` y variables sociales-demográficas (`pp_total_raw`, `pct_free/reduced`, `pct_black/hispanic`, `locale`, `Sector(s)`) contribuyen en menor medida.

Por último, veamos como trabaja este modelo cuando nos referimos a las predicciones.

```python
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

RND = 42

def smape(a, f):
    a = np.array(a)
    f = np.array(f)
    denom = (np.abs(a) + np.abs(f)) / 2
    mask = denom != 0
    return (100.0 / len(a[mask])) * np.sum(np.abs(a[mask] - f[mask]) / denom[mask])


preds_rf = rf_res['preds']
y_true = rf_res['y_true']

test_df_available = 'X_test' in globals() and isinstance(X_test, (pd.DataFrame, pd.Series))
X_test_df = X_test.reset_index(drop=True).copy()


preds = np.array(preds_rf)
truth = np.array(y_true)

mse = mean_squared_error(truth, preds)
rmse = np.sqrt(mse)
mae = mean_absolute_error(truth, preds)
r2 = r2_score(truth, preds)
sm = smape(truth, preds)
print(f"RMSE: {rmse:.3f} | MAE: {mae:.3f} | R2: {r2:.3f} | SMAPE: {sm:.3f}%")
print("N test:", len(truth))

# predicción vs realidad
plt.figure(figsize=(6,6))
plt.scatter(truth, preds, alpha=0.45, s=15)
mn = min(truth.min(), preds.min())
mx = max(truth.max(), preds.max())
plt.plot([mn, mx], [mn, mx], '--', color='k')
plt.xlabel('True engagement_index_mean')
plt.ylabel('Predicted engagement_index_mean')
plt.title('Predicho vs Real (RandomForest)')
plt.tight_layout()
plt.show()

# Boxplot residuales top 5 estados
df_err = X_test_df.copy()
df_err['y_true'] = truth
df_err['pred'] = preds
df_err['residual'] = df_err['y_true'] - df_err['pred']
top_states = df_err.groupby('state')['y_true'].count().nlargest(5).index.tolist()
df_top = df_err[df_err['state'].isin(top_states)]
plt.figure(figsize=(10,5))
sns.boxplot(x='state', y='residual', data=df_top)
plt.axhline(0, color='k', linestyle='--')
plt.title('Residuals por estado (top 5 por n_obs)')
plt.tight_layout()
plt.show()


#tabla resumen por estado
summary = df_err.groupby('state').apply(
    lambda d: pd.Series({
        'n': len(d),
        'true_mean': d['y_true'].mean(),
        'pred_mean': d['pred'].mean(),
        'rmse_state': np.sqrt(mean_squared_error(d['y_true'], d['pred']))
    })
).sort_values('n', ascending=False).reset_index()
display(summary.head(20))

```

El modelo captura algo de señal (R^2 aprox. 0.52) pero comete errores importantes en casos individuales (dado RMSE y SMAPE altos). La diferencia entre RMSE y MAE y el SMAPE elevado nos sugieren la presencia de outliers y heterocedasticidad, es decir, que la varianza de los errores no es constante.

El gráfico de Predicho vs Real muestra que, en general, las predicciones se alinean con la diagonal, pero a medida que aumenta el valor real aparecen desviaciones mayores. Además, los boxplots muestran la dispersión de residuales por estado; estados con cajas desplazadas o con extremos muy grandes indican sesgos o outliers locales.

### Análisis Parte 2 ###

Comenzamos importando las librerias que usamos y fijando semilla. Luego leemos los datos descargados desde kaggle y obtenemos información respecto a sus valores nulos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sklearn.model_selection import StratifiedShuffleSplit


import joblib
RND = 42
```

```python
train_df = pd.read_csv('titanic/train.csv')
test_df  = pd.read_csv('titanic/test.csv')

print("Train shape:", train_df.shape)
print("Test shape: ", test_df.shape)

print("\nNulos en train:")
print(train_df.isnull().sum())

print("\nNulos en test:")
print(test_df.isnull().sum())

display(train_df.head())

print("\nDistribución target (Survived):")
print(train_df['Survived'].value_counts(normalize=True))
```

El análisis muestra que el dataset de entrenamiento tiene 891 filas y el de prueba 418, con valores nulos principalmente ubicados en las columnas Age, Cabin y Embarked. Además, la variable objetivo Survived está desbalanceada, con un 38% de sobrevivientes.
A continuación, creamos la función de preprocesamiento.

```python
def preprocess_data(df, is_train=True):
    df_processed = df.copy()
    df_processed['FamilySize'] = df_processed['SibSp'] + df_processed['Parch'] + 1
    df_processed['IsAlone'] = (df_processed['FamilySize'] == 1).astype(int)
    df_processed['Title'] = df_processed['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
    title_mapping = {
        'Mr': 'Mr', 'Miss': 'Miss', 'Mrs': 'Mrs', 'Master': 'Master',
        'Dr': 'Rare', 'Rev': 'Rare', 'Col': 'Rare', 'Major': 'Rare',
        'Mlle': 'Miss', 'Ms': 'Miss', 'Lady': 'Rare', 'Sir': 'Rare',
        'Mme': 'Mrs', 'Don': 'Rare', 'Capt': 'Rare', 'Countess': 'Rare',
        'Jonkheer': 'Rare', 'Dona': 'Rare'
    }

    df_processed['Title'] = df_processed['Title'].map(title_mapping).fillna('Rare')
    df_processed['Age'] = df_processed['Age'].fillna(df_processed['Age'].median())
    df_processed['Fare'] = df_processed['Fare'].fillna(df_processed['Fare'].median())
    df_processed['Embarked'] = df_processed['Embarked'].fillna(df_processed['Embarked'].mode()[0])
    df_processed['Cabin'] = df_processed['Cabin'].fillna('U')
    df_processed['Cabin'] = df_processed['Cabin'].astype(str).str[0]

    cols_to_drop = ['Name', 'Ticket']
    df_processed.drop([c for c in cols_to_drop if c in df_processed.columns], axis=1, inplace=True, errors='ignore')
    return df_processed

train_p = preprocess_data(train_df)
test_p  = preprocess_data(test_df)
print("Shapes preprocesados:", train_p.shape, test_p.shape)
```

Ahora preparamos matrices con nuestras variables y sincronizamos las columnas de train y test.

```python
train_proc = train_p.copy()
test_proc  = test_p.copy()

y = train_proc['Survived'].copy()
X = train_proc.drop(['Survived','PassengerId'], axis=1, errors='ignore')
test_passenger_ids = test_proc['PassengerId'].copy()
X_test = test_proc.drop(['PassengerId'], axis=1, errors='ignore')

common_cols = X.columns.intersection(X_test.columns)
X = X[common_cols].copy()
X_test = X_test[common_cols].copy()

print("Columnas usadas:", list(common_cols))
print("X shape:", X.shape, " X_test shape:", X_test.shape)
```

Ahora separamos numéricos y categóricos para luego realizar un split en el conjunto de entrenamiento y conseguir el mejor modelo.

```python
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(include=['object','category']).columns.tolist()


print("Num cols:", num_cols)
print("Cat cols:", cat_cols)

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('ohe',  pd.get_dummies)
])

X_enc = pd.get_dummies(X, columns=cat_cols, dummy_na=False)
X_test_enc = pd.get_dummies(X_test, columns=cat_cols, dummy_na=False)

X_enc, X_test_enc = X_enc.align(X_test_enc, join='left', axis=1, fill_value=0)
print("X_enc shape:", X_enc.shape, "X_test_enc shape:", X_test_enc.shape)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_enc, y, test_size=0.2, random_state=RND, stratify=y)

print("Train/val shapes:", X_train.shape, X_val.shape)

models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=RND),
    'Random Forest': RandomForestClassifier(n_estimators=200, random_state=RND),
    'Gradient Boosting': GradientBoostingClassifier(random_state=RND),
    'SVM': SVC(random_state=RND, probability=True)
}

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled   = scaler.transform(X_val)
X_test_scaled  = scaler.transform(X_test_enc)

results = {}
for name, model in models.items():
    if name in ['Logistic Regression', 'SVM']:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_val_scaled)
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_val)
    acc = accuracy_score(y_val, y_pred)
    results[name] = (acc, model)
    print(f"{name}: {acc:.4f}")

best_model_name = max(results, key=lambda k: results[k][0])
best_model = results[best_model_name][1]
print("Mejor modelo:", best_model_name)
```

Notamos que la regresión logística supera por poco a los demás modelos, que también tienen un buen desempeño. Lo que indica que las variables predictoras y la supervivencia son en gran parte lineales y que el modelo se consigue adaptar sin un sobreajuste evidente.

Ahora, aplicaremos validación cruzada sobre este modelo que obtuvims.

```python
sk = StratifiedKFold(n_splits=5, shuffle=True, random_state=RND)
pipe_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(max_iter=5000, random_state=RND))
])

param_grid = {
    'clf__C': [0.01, 0.1, 1, 10, 100],
    'clf__penalty': ['l2'],
    'clf__solver': ['lbfgs', 'saga'],
    'clf__class_weight': [None, 'balanced']
}

grid = GridSearchCV(pipe_lr, param_grid, cv=sk, scoring='accuracy', n_jobs=-1, verbose=1)
grid.fit(X_enc, y)   # ejecutamos CV sobre todo el conjunto de entrenamiento preprocesado

print("Parámetros:", grid.best_params_)
print("Mejor score CV (accuracy):", grid.best_score_)


best_model = grid.best_estimator_
val_pred = best_model.predict(X_val)
val_acc = accuracy_score(y_val, val_pred)
print(f"Accuracy en validation set con el mejor modelo: {val_acc:.4f}")

coefs = best_model.named_steps['clf'].coef_
names = X_enc.columns
coef_imp = np.abs(coefs).mean(axis=0) if coefs.ndim>1 else np.abs(coefs)
coef_df = pd.DataFrame({'feature': names, 'coef_abs': coef_imp}).sort_values('coef_abs', ascending=False)
display(coef_df.head(20))
```

Notamos que `Pclass` es el predictor más fuerte (con mayor coeficiente absoluto), lo cual coincide con la intuición histórica respecto al Titanic (la clase influye mucho en supervivencia). Además, Muchos títulos (`Mr`, `Master`, `Mrs`, `Miss`) aparecen entre los top, lo que refleja que el tratamiento del nombre/título captura señal demográfica importante.

Luego, entrenamos el modelo y creamos el submission.

```python
best_model.fit(np.vstack([X_train_scaled, scaler.transform(X_val)]), np.hstack([y_train, y_val]))
test_preds = best_model.predict(X_test_scaled)

submission = pd.DataFrame({'PassengerId': test_passenger_ids, 'Survived': test_preds})
print(submission.head(10))
submission.to_csv('titanic_predictions.csv', index=False)
print("Saved titanic_predictions.csv")
```

Una vez subido a kaggle, obtuvimos un puntaje final de 0.768 aprox.. Concluyendo en un modelo bastante competente.
